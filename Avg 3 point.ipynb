# Analysis of the Top 100 NBA Three Point Percentage Rankings

# 

## Importing Libraries

```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import matplotlib.pyplot as plt
```

## Fetching URL and Creating a Data Frame

```python
url = "https://www.teamrankings.com/nba/player-stat/three-point-field-goal-percentage"

req = requests.get(url)

sopa = BeautifulSoup(req.text, 'lxml')
```

```python
# Creating Players List
p = sopa.select("td")[1:500:5]
```

```python
p
```

```python
# Cleaning up players list
players = []

for x in p:
    player = x.text.replace('\n', '').replace("'", '')
    
    players.append(player)
    
    
    
    
```

```python
players
```

```python
# Create teams list
t = sopa.select("td")[2:500:5]
```

```python
# Clean up teams list
teams = []

for x in t:
    
    teams.append(x.text)
```

```python
teams
```

```python
# Create position list 
pos = sopa.select("td")[3:500:5]
```

```python
# Clean up position list
position = []

for p in pos:
    
    position.append(p.text)
```

```python
position
```

```python
# Create Value list
v = sopa.select("td")[4:500:5]
```

```python
# Clean up value list
values = []

for x in v:
    
    values.append(x.text)
```

```python
values
```

```python
# Create a data frame from dictionary vaules attached to a variable
p_dict = {"Player": players, "Team": teams, "Position": position, "Value": values}

df = pd.DataFrame(p_dict)

#convert data frame into a CSV file
df.to_csv("NBA_3point_%_list")

```

```python
# Read CSV file
data = pd.read_csv("NBA_3point_%_list")
```

```python
data
```

```python
# Check the position column in my Data Frame
data["Position"]
```

## Analyze number of Players By Position in NBA 3 Point Percentage Data Frame

```python
# Count the number of values in each position
position_counts = data['Position'].value_counts()

# Print the number of players by position
for position, count in position_counts.items():
    print(f"Number of players in {position} position is {count}")

# Create a bar chart of the position counts
fig, ax = plt.subplots(figsize=(12,10))
position_counts.plot(kind='bar', ax=ax)

# Set the chart title and axis labels
plt.title('Number of Players by Position')
plt.xlabel('Position')
plt.ylabel('Number of Players')

# Add visible lines for each number on the y axis
ax.yaxis.grid(True)

# Display the chart
plt.show()
```

### By Far Shooting Guards account for most of the players by position. Centers account for the least amount of Players by position.

## Convert Each Value in the data[Value] column into a Float to find the mean.

## Average Value By Position

```python
# Convert 'Value' to string, then remove the percent sign and convert to float
data['Value'] = data['Value'].astype(str).str.replace('%', '').astype(float)

# Group the dataframe by 'Position' and calculate the mean of 'Value' for each group
position_means = data.groupby('Position')['Value'].mean()

# Sort the position means dictionary in descending order
position_means_sorted = dict(sorted(position_means.items(), key=lambda item: item[1], reverse=True))

# Print the average value by position in descending order
for position, mean in position_means_sorted.items():
    print(f"Average value by {position} position is {mean:.2f}")

# Create a bar chart of the position means in descending order
fig, ax = plt.subplots(figsize=(12,10))
pd.Series(position_means_sorted).plot(kind='barh', ax=ax)

# Set the chart title and axis labels
plt.title('Average Value by Position')
plt.xlabel('Average Percentage')
plt.ylabel('Position')

# Add visible lines for each number on the x axis
ax.xaxis.grid(True)


# Display the chart
plt.show()
```

### Although Centers account for the fewest players by position. They have the highest Average by Position.

## Visualize how many players in this Data Frame are in the Eastern Conference compared to The Western Conference

```python
# Create a list of East conference teams and West conference teams
east_teams = ['Atlanta Hawks', 'Boston Celtics', 'Brooklyn Nets', 'Charlotte Hornets', 'Chicago Bulls', 'Cleveland Cavaliers', 'Detroit Pistons', 'Indiana Pacers', 'Miami Heat', 'Milwaukee Bucks', 'New York Knicks', 'Orlando Magic', 'Philadelphia 76ers', 'Toronto Raptors', 'Washington Wizards']
west_teams = ['Dallas Mavericks', 'Denver Nuggets', 'Golden State Warriors', 'Houston Rockets', 'Los Angeles Clippers', 'Los Angeles Lakers', 'Memphis Grizzlies', 'Minnesota Timberwolves', 'New Orleans Pelicans', 'Oklahoma City Thunder', 'Phoenix Suns', 'Portland Trail Blazers', 'Sacramento Kings', 'San Antonio Spurs', 'Utah Jazz']

# Count the number of East and West conference teams in the dataframe
east_count = data['Team'].isin(east_teams).sum()
west_count = data['Team'].isin(west_teams).sum()

# Set the figure size
plt.figure(figsize=(8, 6))

# Create a pie chart of the team counts by conference
labels = ['East', 'West']
counts = [east_count, west_count]
plt.pie(counts, labels=labels, autopct='%1.1f%%', startangle=90)

# Add player count text to the plot
plt.text(0.5, 1.15, f'East: {east_count} players\nWest: {west_count} players', ha='center', va='center', transform=plt.gca().transAxes)

# Set the chart title and legend
plt.title('NBA Team Counts by Conference')
plt.legend(title='Conference', loc='best')

# Display the chart
plt.show()
```

### The Western Conference acounts for more players in the Data Frame.

## How Many Players From each NBA team are present in the Data Frame

```python
# Count the number of players on each team
team_counts = data['Team'].value_counts()

# Get the teams with the most players
most_players_teams = list(team_counts.head(4).index)

# Print the teams with the most players
print("The teams with the most players are:")
for team in most_players_teams:
    print(f"- {team} ({team_counts[team]} players)")

# Create a bar chart of the team counts
fig, ax = plt.subplots(figsize=(14,12)) # Increase the size of the chart
team_counts.plot(kind='barh', ax=ax) # Set the chart to use the ax instance

# Set the chart title and axis labels
ax.set_title('NBA Team Player Counts') # Use the ax instance to set the title
ax.set_xlabel('Player Count')
ax.set_ylabel('Team')

# Display the chart
plt.show()
```
